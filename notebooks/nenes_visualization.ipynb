```python
from lognorm import Lognorm
l = Lognorm(mu=0.1, sigma=2, N=3000.)

rs = np.logspace(-4, 1, num=100)
plot(rs, l.pdf(rs), '-o')
semilogx()

mids = np.array([np.sqrt(a*b) for a, b in zip(rs[:-1], rs[1:])])
plot(mids, l.pdf(mids), 'kx')
```

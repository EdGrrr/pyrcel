```python
## Model parameters
dt = 1.0 # seconds
dz = 200.0 # meters
z_top = 8e3 # meters

z = np.arange(dz, z_top+1e-1, dz)

## Pick a sounding, and interpolate the virtual temperature data
sounding = read_sounding("data/HPII_2005-06-01-11:00:00.csv")
#sounding = read_sounding("data/87155_2005-06-01-12:00:00.csv")
print sounding['height']
sounding['virtual temperature'] = virtual_temperature(sounding)
ps = parcel_sounding(sounding)

## Trucnate the sounding to usable data
sounding = sounding[sounding['height'] < 9e3]
ps = ps[ps['height'] < 9e3]

## Interpolate the virtual temperature soundings
sounding_tv = interp1d(sounding['height'], sounding['virtual temperature'], 'cubic')
ps_tv = interp1d(ps['height'], ps['virtual temperature'], 'cubic')

# Visual check of soundings
figure(figsize=(10, 4))
subplot(1,2,1)
plot((ps_tv(z)-sounding_tv(z))/sounding_tv(z), z)
vlines([0.], dz, z_top, linestyle='dotted')
title("$\Delta T_v / T_v$")
subplot(1,2,2)
plot(sounding_tv(z), z, 'r', label='$T_v$')
plot(ps_tv(z), z, 'b', label='$T_{v,cloud}$')
legend()
plot_sounding(sounding, ps)

# Compute density profile
sounding['density'] = (sounding['pressure']*100.)/(Rd*sounding['virtual temperature'])
sounding_dens = interp1d(sounding['height'], sounding['density'], 'cubic')
rho = sounding_dens(z)
figure(figsize=(10, 4))
subplot(1,2,1)
plot(sounding_dens(z), z)
title(r'$ \rho $')

subplot(1,2,2)
drho_dz = np.zeros_like(rho)
drho_dz[1:-1] = (1./(2.*dz))*(rho[2:] - rho[:-2])
drho_dz[0] = (1./(2.*dz))*(-rho[2] + 4.*rho[1] - 3.*rho[0])
drho_dz[-1] = (1./(2.*dz))*(rho[-3] - 4.*rho[-2] + 3.*rho[-1])
rho_star = (1./rho)*drho_dz
plot(rho_star, z)
title(r"$\frac{1}{\rho}\;\frac{d\rho}{dz}$")

# Compute water vapor mixing ratio profile
sounding['water vapor'] = epsilon*(sounding['relative humidity']/100.)*es(sounding['temperature'])/(sounding['pressure']*100.)
sounding_wv = interp1d(sounding['height'], sounding['water vapor'], 'cubic')
wv = sounding_wv(z)

figure(figsize=(10,4))
subplot(1,2,1)
plot(wv, z)
title(r'$Q_v$')

subplot(1,2,2)
dwv_dz = np.zeros_like(wv)
dwv_dz[1:-1] = (1./(2.*dz))*(wv[2:] - wv[:-2])
dwv_dz[0] = (1./(2.*dz))*(-wv[2] + 4.*wv[1] - 3.*wv[0])
dwv_dz[-1] = (1./(2.*dz))*(wv[-3] - 4.*wv[-2] + 3.*wv[-1])
plot(dwv_dz, z)
title(r'$\frac{d Q_v}{dz}$')
```

```python
# Try to compute updraft velocity
dtv = (ps_tv(z)-sounding_tv(z))/sounding_tv(z)
Us = []
for i, zi in enumerate(z):
    U = np.sqrt(2*g*trapz(dtv[:i], z[:i]))
    Us.append(U)
Us = np.array(Us)
plot(Us, z)
```

```python
def run_model(u_initial, r_initial, environment, dt=2.0, t_end=60.*25):

    counter = 0
    dtv, rho_star, G, Vc = environment['dtv'], environment['rho_star'], environment['G'], environment['Vc']
    
    U = u_initial.copy()
    R = r_initial.copy()
    
    t = 0
    U_save = []
    R_save = []
    while t < t_end:
        t = t+dt
        
        ## 1) Integrate U forward in time using Euler Forward, with upwind differencing on the
        ##    spatial derivative term. Prescribe U(t, z=0) = 0. At top boundary, if U < 0 (downdraft),
        ##    then let U(t=t+dt, z=z_top) = U(t=t, z=z_top) (leave unchanged)
        U_change = np.zeros_like(U)
        
        # need to compute dU_dz using an upwind scheme
        def upwind(wind, vector, dz, correction=0.):
            '''Upwind difference, computed on interior points of "vector"'''
            upwind_diff = []
            for i in xrange(1, len(vector)-1):
                if wind[i]-correction > 0:
                    d = (wind[i]-correction)*(vector[i] - vector[i-1])
                else:
                    d = (wind[i]-correction)*(vector[i+1] - vector[i])
                upwind_diff.append(d)
            upwind_diff = np.array(upwind_diff)
            return (1./dz)*upwind_diff
        
        U_change[1:-1] = dt*(g*(dtv[1:-1] - R[1:-1]) - upwind(U, U, dz))
        #if t%10==0:
        #    print t, upwind(U, dz), "\n"
        # Top BC
        if U[-1] > 0:
            U_change[-1] = dt*(g*(dtv[-1] - R[-1]) - (U[-1]/dz)*(U[-1] - U[-2]))
        else:
            U_change[-1] = 0
            #U_change[-1] = dt*g*(dtv[-1] - R[-1])
        # Bottom BC
        U = U + U_change    
    
        ## 2) Integrate R forward in time using Euler Forward, with upwind differencing, same as before
        R_change = np.zeros_like(R)
        
        R_change[1:-1] = dt*(U[1:-1]*G[1:-1] + R[1:-1]*Vc*rho_star[1:-1] - upwind(U, R, dz, Vc))
        # Top BC
        if U[-1] >= 0:
            R_change[-1] = dt*(U[-1]*G[-1] + R[-1]*Vc*rho_star[-1] - ((U[-1]-Vc)/dz)*(R[-1] - R[-2]))
        else:
            R_change[-1] = 0
            #R_change[-1] = dt*(((U[-1]-Vc)/dz)*(R[-1] - R[-2]))
            #R_change[-1] = dt*(U[-1]*G[-1] + R[-1]*Vc*rho_star[-1])
        # Bottom BC
        R_change[0] = dt*(R[0]*Vc*rho_star[0] + (Vc/dz)*(R[1] - R[0]))
        #R_change[0] = dt*(R[0]*Vc*rho_star[0] + (1./(2.*dz))*(-R[2] + 4.*R[1] - 3.*R[0]))
        R = R + R_change
        R[R < 0.] = 0.
    
        if t%(60.*5.)==0.: 
            print t/(60.*5.), counter
            U_save.append(U)
            R_save.append(R)
        
        counter += 1
    
    print counter
        
    return U_save, R_save
```

```python
def plot_result(U_save, R_save):
    figure(figsize=(10, 4))
    subplot(1,2,1)
    plot(Us, z, label="t=0")
    for i, U in enumerate(U_save):
        plot(U, z, label="t=%d"%((i+1)*5))
    vlines([0., ], 0, z_top, linestyle='dotted')
    #legend()
    xlabel(r'U (m/s)')
    xlim(-20, 20)
    
    subplot(1,2,2)
    plot(Rs*1000., z)
    for i, R in enumerate(R_save):
        plot(R*1000., z, label="t=%d"%((i+1)*5))
    xlabel(r'R (g/kg)')
    xlim(0, 10)
    
    #subplot(1,3,3)
    #plot(dtv, z)
    #xlabel(r'$\Delta T_v / T_v$')
    #print U[-5:]
```

```python
def upwind(wind, vector, dz, correction=0.):
    '''Upwind difference, computed on interior points of "vector"'''
    upwind_diff = []
    for i in xrange(1, len(vector)-1):
        if wind[i]-correction > 0:
            d = (wind[i]-correction)*(vector[i-1] - vector[i])
        else:
            d = (wind[i]-correction)*(vector[i] - vector[i+1])
        upwind_diff.append(d)
    upwind_diff = np.array(upwind_diff)
    return (1./dz)*upwind_diff

dzz = 1.
dt = 1.
zz = np.arange(0, 1000+dzz, dzz)

x = np.zeros_like(zz)
x[(zz > 400) & (zz < 500)] = 1.0
u = np.ones_like(zz) * -1

subplot(1,2,1)
plot(x, zz)
xlim(0, 2)
subplot(1,2,2)
plot(zz, x)
ylim(0, 2)

for i in xrange(100):
    x[1:-1] = x[1:-1]+upwind(u, x, dzz)
subplot(1,2,1)
plot(x, zz)
subplot(1,2,2)
plot(zz, x)
```

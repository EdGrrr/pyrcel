```python
import numpy as np
from scipy.interpolate import interp1d
from scipy.integrate import trapz, quad
from scipy.special import gamma

from microphysics import *
from pywrfplotUtils import es
from soundings import *

from pylab import *
ion()

def d_dz(u, dz):
    du_dz = np.zeros_like(u)
    du_dz[1:-1] = (1./(2.*dz))*(u[2:] - u[:-2])
    du_dz[0] = (1./(2.*dz))*(-u[2] + 4.*u[1] - 3.*u[0])
    du_dz[-1] = (1./(2.*dz))*(u[-3] - 4.*u[-2] + 3.*u[-1])
    return du_dz
```

```python
cloud_base_z = 1500. # meters
cloud_top_z = 8000. # meters

dz = 100. # meters
z_top = 9.5e3 # meters
z = np.arange(0, z_top+dz, dz)

cloud_base = np.where(z == cloud_base_z)[0]
cloud_top = np.where(z == cloud_top_z)[0]
in_cloud_z = (z > cloud_base_z) & (z < cloud_top_z)
```

#### Set up the sounding ####

```python
sounding = DataFrame({'height': z})

## CONDITION 1S from thesis paper
lapse_rate = 6.8e-3 # K/m


## Set up the thermodynamic/meteorology environment based on a very simple 
## tropical convection example (constant lapse rate)
# Temperature
t = np.zeros_like(z)
t[cloud_base] = 277.2 # K
for i in xrange(cloud_base+1, len(t)):
    t[i] = t[i-1] - lapse_rate*dz
for i in xrange(cloud_base):
    t[cloud_base-1-i] = t[cloud_base-1-i+1] + lapse_rate*dz
sounding['temperature'] = t
    
# Pressure, assuming hydrostatic balance and a locally constant scale height
# derived from the temperature profile
H = 287.*t/g
p0 = 100000.
lp = [np.trapz(1./H[:i], z[:i]) for i in xrange(len(z))]
p = p0*np.exp(-np.array(lp))
pressures = p
dp = pressures[2]-pressures[1]
pressures[0] = pressures[1]-dp
sounding['pressure'] = pressures

# Moisture, assuming constant below cloud
x = np.zeros_like(z)
x[cloud_base] = 0.00603 # kg/kg
x[:cloud_base] = x[cloud_base]
q = np.zeros_like(x[cloud_base:])
q = 0.622*1.003*es(t[cloud_base:]-273.15)/pressures[cloud_base:]
#x[cloud_base:] = q/(1.-q)
x[cloud_base:] = q

# Cloud Droplet Generating Function
G = -d_dz(x, dz)/2.

# Virtual temperature
tv = (1.+0.61*x)*t

# Density
rho = pressures/(Rd*t)
rho_star = (1./rho)*d_dz(rho, dz)

# Cloud virtual temperature. Start at cloud base, integrate the moist adiabatic lapse
# rate vertically
tv_cloud = np.zeros_like(z)
tv_cloud[:cloud_base] = tv[:cloud_base]
for i in xrange(cloud_base-1, len(tv_cloud)-1):
    dp = (pressures[i+1] - pressures[i-1])/2.
    t = tv_cloud[i]/(1.+0.61*x[i])
    dt_dp = 0.5*(gamma_s(t-273.15, pressures[i+1]) + gamma_s(t-273.15, pressures[i-1]))
    t_adv = t+dt_dp*dp
    #t_adv = t+(gamma_s(t-273.15, pressures[i])*dp)
    #print t, t_adv
    tv_cloud[i+1] = (1.+0.61*x[i])*t_adv
tv_cloud[cloud_top:] = tv[cloud_top:]
plot(tv, z)
plot(tv_cloud, z, linestyle='dashed')

# Virtual temperature excess
dtv = (tv_cloud-tv)/tv
dtv[dtv < 0] = 0

dtv_f = interp1d(z, dtv, 'slinear')
rho_star_f = interp1d(z, rho_star, 'slinear')
G_f = interp1d(z, G, 'slinear')

Vc = 5.

environment = {'Vc':Vc, 'rho_star':rho_star, 'dtv':dtv, 'G':G}
```

#### Calculate initial updraft and rain mixing ratio ####

```python
y0, z0 = [0.5, 1e-5], cloud_base_z+dz
def f(t, y, *arg1):
    u_z = (g/y[0])*(dtv_f(t) - y[1])
    #r_z = (y[0]*G_f(t) + y[1]*Vc*rho_star_f(t))/(y[0] - Vc)
    r_z = (y[0]*G_f(t))/y[0]
    #print G_f(t), rho_star_f(t), dtv_f(t)
    return [u_z, r_z]
    
from scipy.integrate import ode
r = ode(f).set_integrator('vode', method='adams', order=12)
#r = ode(der).set_integrator('dop853')

r.set_initial_value(y0, z0)
dt = 1.
out = [y0, ]
zs = [z0, ]
while r.successful() and r.t < cloud_top_z:
    r.integrate(r.t+dt)
    out.append(r.y)
    zs.append(r.t)
    #print r.t, r.y
out = np.array(out)

subplot(1,2,1)
plot(out[:,0], zs)
ylim(0, z_top)
subplot(1,2,2)
plot(out[:,1], zs)
ylim(0, z_top)

u_interp = interp1d(zs, out[:,0], 'slinear')
r_interp = interp1d(zs, out[:,1], 'slinear')
zs_min, zs_max = np.min(zs), np.max(zs)
```

```python
Us = np.zeros_like(z)
Rs = np.zeros_like(z)
for i, zi in enumerate(z):
    if zs_min < zi < zs_max: 
        Us[i] = u_interp(zi)
        Rs[i] = r_interp(zi)
```

### MODEL ###

```python
def upwind(winds, vector, dz):
    '''Upwind difference, computed on interior points of "vector"'''
    if len(winds) == 2:
        wind_n   = winds[0]
        wind_n_1 = winds[1]
        wind_ave = 0.5*(wind_n + wind_n_1)
    if len(winds) == 1:
        wind_ave = winds[0]

    upwind_diff = []
    for i in xrange(1, len(vector)-1):
        if wind_ave[i] > 0:
            d = wind_ave[i]*(vector[i] - vector[i-1])
        else:
            d = wind_ave[i]*(vector[i+1] - vector[i])
        upwind_diff.append(d)

    upwind_diff = np.array(upwind_diff)
    return (1./(dz))*upwind_diff

def run_model(u_initial, r_initial, environment, dt=2.0, t_end=60.*25):

    counter = 0
    dtv, rho_star, G, Vc = environment['dtv'], environment['rho_star'], environment['G'], environment['Vc']
    
    U = u_initial.copy()
    R = r_initial.copy()
    
    t = 0
    U_save = []
    R_save = []
    while t < t_end:
        t = t+dt

        U_next = np.zeros_like(U)
        R_next = np.zeros_like(R)

        ## 1) U
        B = dtv - R
        U_next_sub = []
        for i in xrange(1, len(U)-1):
            if U[i] > 0: 
                d = (U[i]*((1./dt)-(U[i]-U[i-1])/(2.*dz))+(g/4.)*(2*B[i]+B[i-1]+B[i+1]))/((1./dt)+(U[i]-U[i-1])/(2.*dz))
            else:
                d = (U[i]*((1./dt)-(U[i+1]-U[i])/(2.*dz))+(g/4.)*(2*B[i]+B[i-1]+B[i+1]))/((1./dt)+(U[i+1]-U[i])/(2.*dz))
            U_next_sub.append(d)
        U_next[1:-1] = np.array(U_next_sub)
        #print U_next

        U_change = np.zeros_like(U)
        if U[-1] > 0:
            U_change[-1] = dt*(g*(dtv[-1] - R[-1]) - (U[-1]/dz)*(U[-1] - U[-2]))
        else:
            U_change[-1] = 0
        # Bottom BC
        U_change[0] = 0.
        U_next[-1] = U[-1] + U_change[-1]
    
        ## 2) Integrate R forward in time using Euler Forward, with upwind differencing, same as before
        R_change = np.zeros_like(R)
        
        R_change[1:-1] = dt*(U[1:-1]*G[1:-1] + R[1:-1]*Vc*rho_star[1:-1] - upwind([U, U_next], R, dz) + Vc*d_dz(R, dz)[1:-1])
        # Top BC
        if U[-1]-Vc >= 0:
            R_change[-1] = dt*(U[-1]*G[-1] + R[-1]*Vc*rho_star[-1] - ((U[-1]-Vc)/dz)*(R[-1] - R[-2]))
        else:
            R_change[-1] = 0
        # Bottom BC
        R_change[0] = dt*(R[0]*Vc*rho_star[0] + (Vc/dz)*(R[1] - R[0]))
        
        ### ADD UP
        U = U_next[:]
        R = R + R_change
        R[R < 0.] = 0.
    
        if t%(60.*5.)==0.: 
            print t/(60.*5.), counter
            U_save.append(U)
            R_save.append(R)
        
        counter += 1
    
    print counter
        
    return U_save, R_save
```

### Lorenz Attractor test

```python
from assimulo.problem import Explicit_Problem
from assimulo.solvers.sundials import CVode

def f(t, u, sigma, rho, beta):
    sigma, rho, beta = [10., 28., 8./3.]
    x, y, z = u
    dx_dt = sigma*(y-x)
    dy_dt = x*(rho-z) - y
    dz_dt = x*y - beta*z
    return np.array([dx_dt, dy_dt, dz_dt])

def rhs(t, u):
    return f(t, u, *args)
    
t0 = 0.
y0 = [1., 1., 1.]
args = [10., 8./3., 28.]

print rhs(t0, y0)

## Create Assimulo explicit problem
prob = Explicit_Problem(rhs, y0)
solver = CVode(prob)
solver.rtol=1e-6
solver.atol=1e-6

t_end = 100.
t, y = solver.simulate(t_end)

prob = Explicit_Problem(rhs, [y0[0], y0[1], y0[2]+1e-2])
solver = CVode(prob)
solver.rtol=1e-6
solver.atol=1e-6

t2, y2 = solver.simulate(t_end)
```

#### Microphysics and auxiliary functions #####

```python
def d_dz(u, dz):
    du_dz = np.zeros_like(u)
    du_dz[1:-1] = (1./(2.*dz))*(u[2:] - u[:-2])
    du_dz[0] = (1./(2.*dz))*(-u[2] + 4.*u[1] - 3.*u[0])
    du_dz[-1] = (1./(2.*dz))*(u[-3] - 4.*u[-2] + 3.*u[-1])
    return du_dz

def plot_result(Us, Ws, Rs, U_save, W_save, R_save):
    figure(figsize=(14, 4))
    subplot(1,3,1)
    plot(Us, z, label="t=0")
    for i, U in enumerate(U_save):
        plot(U, z, label="t=%d"%((i+1)*5))
    vlines([0., ], 0, z_top, linestyle='dotted')
    #legend()
    xlabel(r'U (m/s)')
    xlim(-20, 20)
    
    subplot(1,3,2)
    plot(Ws*1000., z, label="t=0")
    for i, W in enumerate(W_save):
        plot(W*1000., z, label="t=%d"%((i+1)*5))
    xlabel(r'W (g/kg)')
    legend(loc="lower right")
    xlim(0, 16)
    
    subplot(1,3,3)
    plot(Rs*1000., z)
    for i, R in enumerate(R_save):
        plot(R*1000., z, label="t=%d"%((i+1)*5))
    xlabel(r'R (g/kg)')
    xlim(0, 10)

def upwind(wind, vector, dz, correction=None):
    '''Upwind difference, computed on interior points of "vector"
        correction is a vector like vector'''
    upwind_diff = []
    if None == correction: correction = np.zeros_like(vector)
    for i in xrange(1, len(vector)-1):
        if wind[i]-correction[i] > 0:
            d = (wind[i]-correction[i])*(vector[i] - vector[i-1])
        else:
            d = (wind[i]-correction[i])*(vector[i+1] - vector[i])
        upwind_diff.append(d)
    upwind_diff = np.array(upwind_diff)
    return (1./dz)*upwind_diff

def run_model(u_initial, w_initial, r_initial, environment, 
              dt=2.0, t_end=60.*25, steps=None, write_steps=None):

    counter = 0
    dtv, rho_star, G, rho = environment['dtv'], environment['rho_star'], environment['G'], environment['rho']
    
    U = u_initial.copy()
    W = w_initial.copy()
    R = r_initial.copy()    
    
    t = 0
    U_save = []
    W_save = []
    R_save = []
    P1_save = []
    P2_save = []
    V_save = []
    
    if steps is not None:
        t_end = dt*steps
    if None == write_steps:
        write_steps = int(60.*5./dt)
    
    while t < t_end:
        t = t+dt
        
        # Microphysics at this time step
        V = calc_V(R)
        #V = np.ones_like(R)*10.
        P1 = calc_P1(R, U, G)
        P2 = calc_P2(W, R, rho)
        p = P1 + P2
        
        ## 1) Integrate U forward in time using Euler Forward, with upwind differencing on the
        ##    spatial derivative term. Prescribe U(t, z=0) = 0. At top boundary, if U < 0 (downdraft),
        ##    then let U(t=t+dt, z=z_top) = U(t=t, z=z_top) (leave unchanged)
        U_change = np.zeros_like(U)        
        U_change[1:-1] = dt*(g*(dtv[1:-1] - R[1:-1] - W[1:-1]) - upwind(U, U, dz))
        # Top BC
        if U[-1] >= 0:
            U_change[-1] = dt*(g*(dtv[-1] - R[-1] - W[-1]) - (U[-1]/dz)*(U[-2] - U[-1]))
        else:
            U_change[-1] = 0
            #U_change[-1] = dt*g*(dtv[-1] - R[-1])
        # Bottom BC
        U_change[0] = 0.
        U = U + U_change    
        
        ## 2) Integrate W forward in time using Euler Forward, with upwind differencing, same as before. Reset 
        ##    any negative values at W(t+dt) to 0.
        W_change = np.zeros_like(W)
        W_change[1:-1] = dt*(U[1:-1]*G[1:-1] - p[1:-1] - upwind(U, W, dz))
        # Top BC
        if U[-1] >= 0:
            W_change[-1] = dt*(U[-1]*G[-1] - p[-1] - (U[-1]/dz)*(W[-2] - W[-1]))
        else:
            W_change[-1] = 0
        # Bottom BC
        W_change[0] = dt*(-p[0])
        W = W + W_change
        W[W < 0.] = 0.
    
        ## 3) Integrate R forward in time using Euler Forward, with upwind differencing, same as before,
        dV_dz = d_dz(V, dz)
        R_change = np.zeros_like(R)
        R_change[1:-1] = dt*(R[1:-1]*dV_dz[1:-1] + R[1:-1]*V[1:-1]*rho_star[1:-1] + p[1:-1] - upwind(U, R, dz, V))
        # Top BC
        if U[-1] >= 0:
            R_change[-1] = dt*(R[-1]*dV_dz[-1] + R[-1]*V[-1]*rho_star[-1] + p[-1] - ((U[-1]-V[-1])/dz)*(R[-2] - R[-1]))
        else:
            R_change[-1] = 0
        # Bottom BC
        R_change[0] = dt*(R[0]*dV_dz[0] + R[0]*V[0]*rho_star[0] + (V[0]/dz)*(R[1] - R[0]))
        R = R + R_change
        R[R < 0.] = 0.
    
        #if t%(60.*5.)==0.: 
        if (counter+1)%write_steps==0:
            print t/(60.*5.), counter
            U_save.append(U)
            W_save.append(W)
            R_save.append(R)
            P1_save.append(P1)
            P2_save.append(P2)
            V_save.append(V)
        counter += 1
        
        # If anything is NaN, kill simulation
        if np.any(np.isnan(U)) or np.any(np.isnan(W)) or np.any(np.isnan(R)): 
            print "died after %d steps" % counter
            return U_save, W_save, R_save, (P1_save, P2_save, V_save)
    
    print counter
        
    return U_save, W_save, R_save, (P1_save, P2_save, V_save)
```

```python

g3b = gamma(3.+b)
g4b = gamma(4.+b)
g4 = gamma(4.)

calc_lam = lambda r_z: np.power(np.pi*rho_w*N0/r_z, 0.25)
calc_d = lambda lam: 3.67/lam

calc_sp = lambda w_z, r_z, rho: ((np.pi*N0*a*rho*g3b*w_z)/(4.*calc_lam(r_z)**(3.+b)))
def calc_s(w_z):
    s = np.zeros_like(z)
    s[w_z > Wc] = alpha*(w_z[w_z > Wc] - Wc)
    return s

calc_V = lambda r_z: (a*g4b/g4)/(calc_lam(r_z)**b)

calc_P2 = lambda w_z, r_z, rho: calc_s(w_z) + calc_sp(w_z, r_z, rho)
def calc_P1(r_z, u_z, G):
    p1 = np.zeros_like(r_z)
    cond = (u_z < 0) & (r_z > 0) & (w_z <= 0) ## Descending air containing only raindrops
    p1[cond] = u_z[cond]*G[cond]
    return p1

#plot(calc_sp(w_z, r_z, rho), z)
#plot(calc_s(w_z), z)
#figure()
#plot(calc_V(r_z), z)
```

```python
cloud_base_z = 1500. # meters
cloud_top_z = 8500. # meters
dz = 250. # meters
z_top = 9.5e3 # meters
z = np.arange(0, z_top+dz, dz)
in_cloud_z = (z >= cloud_base_z) & (z <= cloud_top_z)

Wc = 3.0/1000. # kg/kg 
alpha = 4e-5 # 1/s
a = (1500.) * 0.1 # cm**(1/2)/s -> m**(1/2)/s 
b = 0.5 # unitless
N0 = 80.*1e5 # 1/(g cm) -> 1/(kg m)
rho_w = 1e3 # kg/m**3

## Read in cloud water data
w_file = open("W.txt", "r")
wzs, ws = [], []
for line in w_file.readlines():
    w, wz = map(float, line.strip().split(", "))
    ws.append(w)
    wzs.append(wz)
w_file.close()

ws = np.array(ws)
ws[ws <= 0.] = 0.0
ws[z < cloud_base_z] = 0.0
ws[z > cloud_top_z] = 0.0

wzs = np.array(wzs)
wzs[0] = 0.

w = DataFrame({"height": wzs*1e3, "w": ws})
w_interp_fcn = interp1d(w['height'], w['w'], 'slinear')
w_upper = np.poly1d(np.polyfit(w['height'], w['w'], 2))
#def w_interp(z):
#    new_z = []
#    for i in xrange(len(z)):
#        if z[i] < cloud_base_z: new_z.append(0.)
#        if z[i] >= cloud_top_z: new_z.append(0.)
#        else: new_z.append(w_interp_fcn(z[i]))
#    return np.array(new_z)
wz = np.zeros_like(z)
wz[in_cloud_z] = w_upper(z[in_cloud_z])
wz[wz < 0] = 0

## Read in updraft data
u_file = open("U.txt", "r")
uzs, us = [], []
for line in u_file.readlines():
    u, uz = map(float, line.strip().split(", "))
    us.append(u)
    uzs.append(uz)
u_file.close()

us = np.array(us)
us[us < 0] = 0.
#us[z < cloud_base_z] = 0.
#us[z > cloud_top_z] = 0.0

uzs = np.array(uzs)
uzs[0] = 0.

u = DataFrame({"height": uzs*1e3, "u": us})
plot(u["u"], u["height"])
u_interp = interp1d(u['height'], u['u'], 'slinear')
u_upper = np.poly1d(np.polyfit(u['height'], u['u'], 2))
uz = np.zeros_like(z)
uz[in_cloud_z] = u_interp(z[in_cloud_z])

subplot(1,2,1)
plot(uz, z)
xlabel(r'U (m/s)')
subplot(1,2,2)
plot(wz, z)
xlabel(r'W (g/kg)')
figure()

plot(u_upper(np.arange(0, z_top+0.1, 0.1)), np.arange(0, z_top+0.1, 0.1))
```
